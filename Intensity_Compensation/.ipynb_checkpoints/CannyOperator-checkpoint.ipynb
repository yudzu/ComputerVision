## Импорт необходимых библиотек

```python
import matplotlib.pyplot as plt 
import numpy as np
import cv2
```

## Загрузка изображения

```python
img = cv2.imread("dull-pink-flowers.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.rcParams['figure.figsize'] = [10, 10]
plt.title("Original image")
plt.imshow(img)
plt.show()
```

## Гистограмма интенсивности оригинального изображения

```python
colors = ('r', 'g', 'b')

for i, color in enumerate(colors): 
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)

plt.title("Original image histogram")
plt.show()
```

## Реализация функции компенсации интенсивности канала

```python
def channel_intensity_compensation(channel):
    """
    Осуществляет компенсацию интенсивности канала изображения
    
    :param channel: канал изображения, у которого необходимо скомпенсировать интенсивность
    
    :return compensated_channel: канал изображения со скомпенсированной интенсивностью
    """
    max_value = np.amax(channel)
    min_value = np.amin(channel)
    ratio = 255 / (max_value - min_value)
    compensated_channel = np.zeros(channel.shape, dtype='uint8')
    
    for i in range(channel.shape[0]):
        for j in range(channel.shape[1]):
            compensated_channel[i][j] = (channel[i][j] - min_value) * ratio
            
    return compensated_channel

r, g, b = cv2.split(img)
r = channel_intensity_compensation(r)
g = channel_intensity_compensation(g)
b = channel_intensity_compensation(b)

compensated_img = cv2.merge((r, g, b))

plt.title("Сompensated image")
plt.imshow(compensated_img)
plt.show()
```

## Гистограмма интесивности изображения после компенсации

```python
for i, color in enumerate(colors): 
    hist = cv2.calcHist([compensated_img], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)

plt.title("Сompensated image histogram")
plt.show()
```

## Нахождение среднего градиента и порогов для оператора Кэнни

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

gradient_img = cv2.Sobel(grayscale_img, cv2.CV_8U, dx=1, dy=1)
mean_gradient = (np.amax(gradient_img) + np.amin(gradient_img)) / 2
print(mean_gradient)

upper_bound = 4 * mean_gradient / 3
lower_bound = 2 * mean_gradient / 3
print("Upper bound: ", upper_bound)
print("Lower bound: ", lower_bound)
```

## Окончательный результат

```python
canny_img = cv2.Canny(img, lower_bound, upper_bound)

plt.imshow(canny_img)
plt.title("Canny image")
plt.show()
```
