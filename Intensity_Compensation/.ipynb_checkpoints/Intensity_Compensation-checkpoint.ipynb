```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("dull-pink-flowers.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.rcParams['figure.figsize'] = [10, 10]
plt.title("Original image")
plt.imshow(img)
plt.show()
```

```python
colors = ('r', 'g', 'b')

for i, color in enumerate(colors): 
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)

plt.title("Original image histogram")
plt.show()
```

```python
def channel_intensity_compensation(channel):
    """
        
    
    :param channel:  ,     
    
    :return compensated_channel:     
    """
    max_value = np.amax(channel)
    min_value = np.amin(channel)
    ratio = 255 / (max_value - min_value)
    compensated_channel = np.zeros(channel.shape, dtype='uint8')
    
    for i in range(channel.shape[0]):
        for j in range(channel.shape[1]):
            compensated_channel[i][j] = (channel[i][j] - min_value) * ratio
            
    return compensated_channel

r, g, b = cv2.split(img)
r = channel_intensity_compensation(r)
g = channel_intensity_compensation(g)
b = channel_intensity_compensation(b)

compensated_img = cv2.merge((r, g, b))

plt.title("ompensated image")
plt.imshow(compensated_img)
plt.show()
```

```python
for i, color in enumerate(colors): 
    hist = cv2.calcHist([compensated_img], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)

plt.title("ompensated image histogram")
plt.show()
```
