```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("dull-pink-flowers.jpg", cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
plt.title("Original image")
plt.show()
```

```python
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist)
plt.title("Original image histogram")
plt.show()
```

```python
max_value = np.amax(img)
min_value = np.amin(img)
diff = max_value - min_value
ratio = 255 / (max_value - min_value)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img[i][j] = (img[i][j] - min_value) * ratio

plt.imshow(img, cmap='gray')
plt.title("ompensated image")
plt.show()
```

```python
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist)
plt.title("ompensated image histogram")
plt.show()
```
